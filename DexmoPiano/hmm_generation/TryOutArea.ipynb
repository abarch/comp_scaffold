```python
from hmmlearn import hmm
import pandas as pd
import os
import numpy as np
import pickle
import math
import matplotlib.pyplot as plt
```

```python
def get_sequence(filename, X, lengths):
    file = pd.read_csv(filename)
    for i, row in file.iterrows():
        X.append(list(row[4:]))
    lengths.append(len(file))
    return X, lengths
```

```python
X = []
lengths = []
for filename in os.listdir("single note"):
    X, lengths = get_sequence("single note" + "/" + filename, X, lengths)
X = np.array(X)

#print(X[:,1])
pitch_right = [float('%.5f' % elem) for elem in X[:,2]]
#print(pitch_right)
anzahl = 0
for i in range(66):
    pitch_right.remove(0.0)#"""
"""for elem in pitch_right:
    if elem > 0.2:
        while elem in pitch_right:
            print("remove")
            pitch_right.remove(elem)"""
#print(pitch_right)
print("anzahl", anzahl)
plt.xlim([0,0.1])
plt.hist(pitch_right, bins = 100)
plt.show()
print("mean", np.mean(pitch_right))
print("median", np.median(pitch_right))
print("max", np.max(pitch_right))
```
